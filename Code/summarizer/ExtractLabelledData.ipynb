```python
import os
import csv
import pandas as pd
class ExtractLabelledData(object):

    categories = ['purpose','description','scope','requirements','assumptions and dependencies','architecture','users']
    path = os.getcwd()+'/../Data'

    def __init__(self):
        print('Extracting Data...')
        self.readData(self.categories)

    def readData(self,categories):
        df = pd.read_csv(self.path+'/labelled_dataset/labelled_data.csv')
        df = df.fillna('')
        display(df)
        self.processData(df,categories)

    def processData(self,df, categories):
        content = {}
        for index,row in df.iterrows():
            for category in categories:
                content[category] = row[category]
            self.writeData(row['doc'], content, categories)

    def writeData(self,srs, content, categories):
        dest = self.path+'summarized_data/content'
        for category in categories:
            write_path = self.path+'/summarized_data/content/'+category+'/'+srs[3:]+'.txt'
            with open(write_path, 'w+') as output:
                output.write(str(content[category]))
```
